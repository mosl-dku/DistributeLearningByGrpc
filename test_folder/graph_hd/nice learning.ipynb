```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
file_name = [['case2_10(loss).csv', 'case2_10(validation, result).csv'],
             ['case3_10(loss).csv', 'case3_10(validation, result).csv']]

file_name
```

```python
data = [[], []]

for i in range(2):
    for j in range(2):
        data[i].append(pd.read_csv(file_name[i][j], header = None))
    
```

```python
train_epoch = []
for i in range(2):
    train_epoch.append(data[i][0].iloc[:, :100])     
```

```python
validation_epoch = []
for i in range(2):
    validation_epoch.append(data[i][1].iloc[:, :100])        
```

```python
# test_data epoch 10당 뽑아내기
'''
'''
```

```python
train_epoch_mean = []

for i in range(2):
    train_epoch_mean.append(train_epoch[i].mean())
```

```python
validation_epoch_mean = []
for i in range(2):
    validation_epoch_mean.append(validation_epoch[i].mean())
```

```python
epoch = 100

plt.figure(figsize=(7,7))
plt.ylim(0, 500)

#plt.subplot(121)
#plt_title = 'Train loss & Validation loss in Case 2'
#plt.title(plt_title, fontsize=15)
plt.ylim(0, 500)
plt.xlabel('Epoch', fontsize=15)
plt.ylabel('Loss(mse)', fontsize=15)
plt.plot(range(epoch), train_epoch_mean[1], label='train_loss')
plt.plot(range(epoch), validation_epoch_mean[1], label='val_loss')
plt.rc('legend', fontsize=15)
plt.legend()
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.savefig("fig3_case3_train.pdf")

# plt.subplot(122)
# plt_title = 'Train loss & Validation loss in Case 3'
# plt.title(plt_title, fontsize=15)
# plt.ylim(0, 500)
# plt.xlabel('Epoch', fontsize=12)
# plt.ylabel('Loss(mse)', fontsize=12)
# plt.plot(range(epoch), train_epoch_mean[1], label='train_loss')
# plt.plot(range(epoch), validation_epoch_mean[1], label='val_loss')
# plt.rc('legend', fontsize=12)
# plt.legend()
# plt.xticks(fontsize=13)
# plt.yticks(fontsize=13)


# plt.title(plt_title, fontsize=15)


# plt.plot(range(epoch), train_data_mean[0], label='2')
# plt.plot(range(epoch), train_data_mean[1], label='5')
# plt.plot(range(epoch), train_data_mean[2], label='10')

# plt.xlabel('Epoch', fontsize=12)
# plt.ylabel('Loss(mse)', fontsize=12)
# plt.rc('legend', fontsize=12)
# plt.legend()

# plt.xticks(fontsize=13)
# plt.yticks(fontsize=13)



plt.savefig("fig3.pdf")
#plt.show()
#plt.savefig(plt_title+'.png')
```
