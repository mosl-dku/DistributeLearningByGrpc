```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
file_name = [['case1_2(validation, result).csv', 'case1_5(validation, result).csv', 'case1_10(validation, result).csv'],
             ['case2_2(validation, result).csv', 'case2_5(validation, result).csv', 'case2_10(validation, result).csv'],
             ['case3_2(validation, result)(again).csv', 'case3_5(validation, result)(again).csv', 'case3_10(validation, result).csv']]
file_name
```

```python
data = [[], [], []]

for i in range(3):
    for j in range(3):
        data[i].append(pd.read_csv(file_name[i][j], header = None))
    
```

```python
train_data = [[], [], []]
for i in range(3):
    for j in range(3):
        train_data[i].append(data[i][j].iloc[:, :100])
```

```python
test_data = [[], [], []]
for i in range(3):
    for j in range(3):
        test_data[i].append(data[i][j].iloc[:, 100:])
```

```python
# test_data epoch 10당 뽑아내기
'''
'''
```

```python
train_data_mean = [[], [], []]
for i in range(3):
    for j in range(3):
        train_data_mean[i].append(train_data[i][j].mean())
```

```python
test_data_mean = [[], [], []]
for i in range(3):
    for j in range(3):
        test_data_mean[i].append(test_data[i][j].mean())
```

```python
df_test_loss = pd.DataFrame({'2'  : [test_data_mean[i][0][102] for i in range(3)],
                             '5'  : [test_data_mean[i][1][102] for i in range(3)],
                             '10' : [test_data_mean[i][2][102] for i in range(3)]},
                             index = ['Case1', 'Case2', 'Case3'])
```

```python
plt_title = 'Test loss of number of remote server by Case'
ax = df_test_loss.plot(kind='bar', figsize=(7, 7), legend=True, rot=0)
#ax.set_title(plt_title, fontsize=16)
ax.set_ylabel('mean deviation', fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.ylim(0, 15)

for p in ax.patches:
    left, bottom, width, height = p.get_bbox().bounds
    ax.annotate("%.1f"%(height), (left+width/2, height*1.01), ha='center')

ax.grid()
#plt.rcParams.update({'font.size': 10})
#plt.legend(title='remote')
plt.rc('legend', fontsize=13)
#ax.legend(loc='lower right')

ax.legend()

plt.savefig("fig2(again).pdf")
```
