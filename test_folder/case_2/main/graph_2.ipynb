```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
file_name = ['case2_2.csv', 'case2_5.csv', 'case2_10.csv',]
```

```python
data = []
for i in range(3):
    data.append(pd.read_csv(file_name[i], header = None))
```

```python
train_data = []
for i in range(3):
    train_data.append(data[i].iloc[:, :100])
```

```python
test_data = []
for i in range(3):
    test_data.append(data[i].iloc[:, 100:])
```

```python
# test_data epoch 10당 뽑아내기
'''
'''
```

```python
train_data_mean = []
for i in range(3):
    train_data_mean.append(train_data[i].mean())
```

```python
test_data_mean = []
for i in range(3):
    test_data_mean.append(test_data[i].mean())
```

```python
test_data_max = []
test_data_min = []

for i in range(3):
    test_data_max.append(test_data[i].max())
    test_data_min.append(test_data[i].min())
```

```python
epoch = 100
plt_title = 'Case2_Train processing'

plt.figure(figsize=(6,6))
plt.title(plt_title, fontsize=15)

plt.ylim(0, 1000)
plt.plot(range(epoch), train_data_mean[0], label='2')
plt.plot(range(epoch), train_data_mean[1], label='5')
plt.plot(range(epoch), train_data_mean[2], label='10')

plt.xlabel('Epoch', fontsize=12)
plt.ylabel('Loss(mse)', fontsize=12)
plt.rc('legend', fontsize=12)
plt.legend()

plt.xticks(fontsize=13)
plt.yticks(fontsize=13)

plt.savefig("Case2_Train loss.pdf")
#plt.show()
#plt.savefig(plt_title+'.png')
```

```python
plt_title = 'Case2_Test loss'

#x = [1, 3, 5, 7]
#x2 = [1.5, 3.5, 5.5, 7.5]y
#y1 = [test_data_mean[0][100], test_data_mean[1][100], test_data_mean[2][100], test_data_mean[3][100]] 
y2 = [test_data_mean[0][102], test_data_mean[1][102], test_data_mean[2][102]] 
x = range(3)
x_name = ['2', '5', '10']

plt.figure(figsize=(6,6))
plt.title(plt_title, fontsize=15)
plt.xticks(x, x_name)

plt.bar(x, y2, width=0.5)

plt.xlabel('Number of remote servers (1/n data per server)', fontsize=12)
plt.ylabel('mean deviation', fontsize=12)

plt.xticks(fontsize=13)
plt.yticks(fontsize=13)

plt.savefig("Case2_Test loss.pdf")

print("2 max: ", test_data_max[0][102])
print("5 max: ", test_data_max[1][102])
print("10 max: ", test_data_max[2][102])


print("2 min: ", test_data_min[0][102])
print("5 min: ", test_data_min[1][102])
print("10 min: ", test_data_min[2][102])
```
