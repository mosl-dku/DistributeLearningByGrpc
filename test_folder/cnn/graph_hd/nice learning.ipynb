```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
file_name = [['cnn_case1(acc).csv', 'cnn_case1(loss).csv'], 
             ['cnn_case2_dis(acc).csv', 'cnn_case2_dis(loss).csv'],
             ['cnn_case2_nodis(acc).csv', 'cnn_case2_nodis(loss).csv'], 
             ['cnn_case3_dis(acc).csv', 'cnn_case3_dis(loss).csv'], 
             ['cnn_case3_nodis(acc).csv', 'cnn_case3_nodis(loss).csv']]

file_name
```

```python
data = [[], [], [], [], []]

for i in range(5):
    for j in range(2):
        data[i].append(pd.read_csv(file_name[i][j], header = None))
    
```

```python
train_epoch = [[], [], [], [], []]

for i in range(5):
    for j in range(2):
        train_epoch[i].append(data[i][j].mean())
```
