```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
file_name = 'local_10.csv'
plt_title = 'Undefine_10'

data = pd.read_csv(file_name, header = None)
```

```python
data_start = 2
data_end   = 57
epoch      = 100

def clear(d:pd.DataFrame):
    rd = pd.DataFrame()
    for i in range(epoch):
        temp = (d.iloc[:, i]).sort_values()
        rd.insert(i, i, value=list(temp)[data_start:data_end])
    return rd
        
data_clear = clear(data)
```

```python
data_min  = data_clear.min()
data_max  = data_clear.max()
data_mean = data_clear.mean()
```

```python
plt.title(plt_title)

plt.plot(range(epoch), data_mean, label='mean')
plt.plot(range(epoch), data_min,  label='min')
plt.plot(range(epoch), data_max,  label='max')

plt.xlabel('Epoch')
plt.ylabel('Loss(mse)')
plt.legend()

#plt.show()
plt.savefig(plt_title+'.png')
```
