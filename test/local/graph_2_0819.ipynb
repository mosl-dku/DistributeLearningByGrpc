```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
file_name = ['local_1.csv', 'local_2.csv', 'local_5.csv', 'local_10.csv']
```

```python
data = []
for i in range(4):
    data.append(pd.read_csv(file_name[i], header = None))
```

```python
train_data = []
for i in range(4):
    train_data.append(data[i].iloc[:, :100])
```

```python
test_data = []
for i in range(4):
    test_data.append(data[i].iloc[:, 100:])
```

```python
# test_data epoch 10당 뽑아내기
'''
'''
```

```python
train_data_mean = []
for i in range(4):
    train_data_mean.append(train_data[i].mean())
```

```python
test_data_mean = []
for i in range(4):
    test_data_mean.append(test_data[i].mean())
```

```python
epoch = 100
plt_title = 'Train processing'

plt.figure(figsize=(6,6))
plt.title(plt_title, fontsize=15)

plt.ylim(0, 1000)
plt.plot(range(epoch), train_data_mean[0], label='non split')
#plt.plot(range(epoch), train_data_mean[1], label='1/2')
#plt.plot(range(epoch), train_data_mean[2], label='1/5')
plt.plot(range(epoch), train_data_mean[3], label='1/10')

plt.xlabel('Epoch', fontsize=12)
plt.ylabel('Loss(mse)', fontsize=12)
plt.rc('legend', fontsize=12)
plt.legend()

plt.xticks(fontsize=13)
plt.yticks(fontsize=13)

plt.savefig("Local_1_10.pdf")
#plt.show()
#plt.savefig(plt_title+'.png')
```

```python
plt_title = 'Test loss'

#x = [1, 3, 5, 7]
#x2 = [1.5, 3.5, 5.5, 7.5]y
#y1 = [test_data_mean[0][100], test_data_mean[1][100], test_data_mean[2][100], test_data_mean[3][100]] 
y2 = [test_data_mean[0][102], test_data_mean[1][102], test_data_mean[2][102], test_data_mean[3][102]] 
x = range(4)
x_name = ['1', '1/2', '1/5', '1/10']

plt.figure(figsize=(7,7))
#plt.title(plt_title, fontsize=15)
plt.xticks(x, x_name)

plt.ylim(0, 15)
plt.grid()
plt.bar(x, y2, width=0.5)

for i, v in enumerate(x):
    plt.text(v, y2[i], round(y2[i], 2),       
             fontsize = 12, 
             color='black',
             horizontalalignment='center',
             verticalalignment='bottom') 

plt.xlabel('Volume of data split', fontsize=15)
plt.ylabel('mean deviation', fontsize=15)

plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

plt.savefig("Local_split_loss.pdf")
```
